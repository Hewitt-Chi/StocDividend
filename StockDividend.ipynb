```python
Seasons = 4
DividendYield = 0.63
StockValue = 53.5
OwnStock = 0
Year,Stock = input("Year, Stock").split()           
TotalTime = int(Year) * Seasons
for i in range(TotalTime) : 
    OwnStock += int(Stock)
    for j in range(3) :
        DividendStock = (OwnStock * DividendYield) / StockValue
        print(f"Dividend Of Every Times : {DividendStock}" + f", Month : {j + 1}")
        OwnStock += DividendStock 
        print(f"After Dividend Stock Count : {OwnStock}")

print(f"Total Stock Count : {OwnStock}")
TotalStockValue = OwnStock * StockValue
print(f"Total Stock Value : {TotalStockValue}")
DividendOfTimes = (OwnStock * DividendYield) / StockValue
print(f"Dividend of times : {DividendOfTimes}") 
print(f"Dividend Value : {DividendOfTimes * StockValue}")
```
